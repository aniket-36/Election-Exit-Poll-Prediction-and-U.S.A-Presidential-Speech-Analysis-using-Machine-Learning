# 2.2) Remove all the stopwords from the three speeches. Show the word count before and after the removal of stopwords. Show a sample sentence after the removal of stopwords.


```python
## Converting to Lower Case
raw1 = inaugural.raw("1941-Roosevelt.txt").lower()
raw1
```

```python
## Cleaning the special characters (/@$: etc)
## from Regular Expressions package
## re.sub() function for string substitution using regular expressions
import re
raw1 = re.sub("[^a-zA-Z]"," ",str(raw1))

raw1
```

```python
## Tokenization -  Splitting the text files into words
tokens1 = nltk.word_tokenize(raw1)
print(tokens1)
```

```python
## Stop words -  Removing words that are not useful or Removing meaningless words 
from nltk.corpus import stopwords
en_stop = stopwords.words('english')
stopped_tokens1 = [i for i in tokens1 if not i in en_stop]
```

# 2.3) Which word occurs the most number of times in his inaugural address for each president? Mention the top three words. (after removing the stopwords)


```python
## Stemming Using Lemmatizer -  Stem the words to its root word
# Roosevelt

from nltk import WordNetLemmatizer
lt1 = nltk.WordNetLemmatizer()
texts1 = [lt1.lemmatize(i) for i in stopped_tokens1]
```

```python
## Top 3 frequency occuring words
Roosevelt = nltk.FreqDist(texts1).most_common(3)
Roosevelt
```

```python
## Plotting the top 3 words occuring frequently
nltk.FreqDist(texts1).plot(3);
```

```python
import matplotlib.pyplot as plt
from nltk.draw.dispersion import dispersion_plot
plt.figure(figsize=(8, 6))
words_used=["nation","life","known"]
dispersion_plot(texts1, words_used, ignore_case=True, title='Dispersion Plot')
```

```python
# Kennedy

from nltk import WordNetLemmatizer
lt2 = nltk.WordNetLemmatizer()
texts2 = [lt2.lemmatize(i) for i in stopped_tokens2]
```

```python
## Top 3 frequency occuring words
Kennedy = nltk.FreqDist(texts2).most_common(3)
Kennedy
```

```python
nltk.FreqDist(texts2).plot(3);
```

```python
plt.figure(figsize=(8, 6))
words_used=["let","u","world"]
dispersion_plot(texts2, words_used, ignore_case=True, title='Dispersion Plot')
```

```python
# Nixon

from nltk import WordNetLemmatizer
lt3 = nltk.WordNetLemmatizer()
texts3 = [lt3.lemmatize(i) for i in stopped_tokens3]
```

```python
## Top 3 frequency occuring words
Nixon = nltk.FreqDist(texts3).most_common(3)
Nixon
```

```python
nltk.FreqDist(texts3).plot(3);
```

```python
plt.figure(figsize=(8, 6))
words_used=["u","let","america"]
dispersion_plot(texts3, words_used, ignore_case=True, title='Dispersion Plot')
```

# 2.4) Plot the word cloud of each of the three speeches. (after removing the stopwords)


```python
## Creating WordCloud Roosevelt
from wordcloud import WordCloud , STOPWORDS 

from wordcloud import WordCloud, STOPWORDS

new_words = set(STOPWORDS)


plt.subplots(figsize=(10,10))
wordcloud = WordCloud(background_color='white',max_words=50,
                            width=1200,stopwords=new_words ,
                            height=1000).generate(" ".join(texts1))

plt.title("MOST USED WORDS",fontsize=20)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
